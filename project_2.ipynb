### Extract Json data

```python
def extract_json(row):
    try:
        parsed_data = json.loads(row)
        return {key: str(value) for key, value in parsed_data.items()}
    except json.JSONDecodeError:
        return {}

# Flatten all JSON fields into separate columns
extracted_data = df1['data'].apply(extract_json).apply(pd.Series)

data1 = pd.concat([df1.drop(columns=['data']), extracted_data], axis=1)
data1.head()
```
